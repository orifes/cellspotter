```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tifffile
```

```python
im = cv2.imread('2.jpg')

im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
# im_gray = cv2.GaussianBlur(im_gray, (1,1), 500)
```

```python
kernel = np.ones((3,3),np.uint8)
eim = cv2.dilate(im_gray, kernel, iterations=3)
plt.imshow(eim, cmap='gray')
cv2.imwrite("bd.jpeg",eim)
```

```python
ret, thresh = cv2.threshold(eim, 170, 255, cv2.THRESH_BINARY)
con, heir = cv2.findContours(image=thresh, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
im_copy = im.copy()
cv2.drawContours(image=im_copy, contours=con, contourIdx=-1, color=(0,255,0), thickness=1)
plt.imshow(im_copy)
cv2.imwrite("bb.jpeg",im_copy)
```

```python
im = cv2.imread("test2.jpg")
plt.imshow(im)
```

```python
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(im_gray, 10, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
con, heir = cv2.findContours(image=thresh, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
im_copy = im.copy()
con = [c for c in con if cv2.contourArea(c)> 50]
cv2.drawContours(image=im_copy, contours=con, contourIdx=-1, color=(0,255,0), thickness=3)
plt.imshow(im_copy)
cv2.imwrite('p.jpg', im_copy)
```

```python
kernel = np.ones((3,3),np.uint8)
eim = cv2.erode(im_gray, kernel, iterations=3)
plt.imshow(eim, cmap='gray')
con, heir = cv2.findContours(image=eim, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
im_copy = eim.copy()
cv2.drawContours(image=im_copy, contours=con, contourIdx=-1, color=(0,0,255), thickness=3)
plt.imshow(im_copy)
cv2.imwrite('p.jpg', im_copy)
```
