```python
i = 3
ch = raw_im[i, : , :]
vmin,vmax = ranges[2*i], ranges[2*i+1]
print(vmin, vmax)
cmap =matplotlib.colors.ListedColormap((luts[i]/256).T.reshape(-1, 3))
mn = matplotlib.colors.Normalize(vmin=vmin, vmax=vmax)
fig = plt.figure(frameon=False)
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(ch, norm=mn, cmap=cmap, aspect='auto')
fig.savefig("test.png",bbox_inches='tight', dpi=100)
new_im = fig.canvas.
# plt.show()
```
