```python
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('520.jpeg')
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im_gray = cv2.GaussianBlur(im_gray, (21,21), 500)
```

```python
ret, thresh = cv2.threshold(im_gray, 150, 255, cv2.THRESH_BINARY)
con, heir = cv2.findContours(image=thresh, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_SIMPLE)
im_copy = im.copy()
cv2.drawContours(image=im_copy, contours=con, contourIdx=-1, color=(0,255,0), thickness=1)
plt.imshow(im_copy)
```
